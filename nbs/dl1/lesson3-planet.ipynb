## Multiclassification

Contrary to the pets dataset studied in last lesson, here each picture can have multiple labels. If we take a look at the csv file containing the labels (in 'train_v2.csv' here) we see that each `image_name` is associated to several `tags` separated by spaces.

```
df = pd.read_csv(path/'train_v2.csv')
df.head()
```

```
df.tail()
```

To put this in a `DataBunch` while using the [**data block API**](https://docs.fast.ai/data_block.html#The-data-block-API), we need to use `ImageList` (and not `ImageDataBunch`). This will make sure the model has the proper loss function to deal with **multiple classes**.

We define the transformations first, having into account the kind of images we have. By default we apply a random horizontal flip, and rotations. We add these:
* `flip_vert`:  images can be vertically flipped or rotated by 90 degrees
* `max_lighting`:  lightning and contrast changes will be applied with the indicated probability.
* `max_zoom`:  adjust the maximum zoom applied
* `max_warp`:  we disable warp because images are always taken with the same angle

```
tfms = get_transforms(flip_vert=True, max_lighting=0.1, max_zoom=1.05, max_warp=0.)
```

Now we follow the usual steps detailed in the data block API:
1. Provide inputs
2. Split the data
3. Label the inputs
4. Apply transformations
5. Convert to a DataBunch
6. Apply normalization

We use parentheses around the data block pipeline below, so that we can use a multiline statement without needing to add '\\'.

We divide the whole process into two. First we get the label lists:

```
np.random.seed(42)
src = (ImageList.from_csv(path, 'train_v2.csv', folder='train-jpg', suffix='.jpg')
       .split_by_rand_pct(0.2)
       .label_from_df(cols='tags', label_delim=' '))
```

```
src
```

Then we create the normalized ImageDataBunch:

```
data = (src.transform(tfms, size=128)
        .databunch().normalize(imagenet_stats))
data
```

The DataBunch's `show_batch` method still works, and show us the different labels separated by `;`.

```
data.show_batch(rows=3, figsize=(12,9))
```

To create a `Learner` we use the same function as in lesson 1. Our base architecture is resnet50 again, but the metrics are a little bit different: we use `accuracy_thresh` instead of `accuracy`. In lesson 1, we determined the prediction for a given class by picking the biggest final activation as the result, but here, each activation can be 0. or 1. `accuracy_thresh` selects the ones that are above a certain threshold (0.5 by default) and compares them to the ground truth. A 0.2 threshold works pretty well always.

As for Fbeta, it's the metric that was used by Kaggle on this competition. Go [here](https://en.wikipedia.org/wiki/F1_score) for more details.

We use `partial` (Python 3) for partial function application, basically when we want to apply always the same parameters to another function (wrapping).

```
arch = models.resnet50
```

```
help(accuracy_thresh)
```

```
acc_02 = partial(accuracy_thresh, thresh=0.2)
f_score = partial(fbeta, thresh=0.2)
learn = cnn_learner(data, arch, metrics=[acc_02, f_score])
```

Fastai chooses the loss function for us depending on the model. We can easily get that info:

```
learn.loss_func
```

We use the LR Finder to pick a good learning rate.

```
learn.lr_find()
```

```
learn.recorder.plot()
```

We need to find the point with the steepest slope. So we can choose a leaning rate somewhere around 1e-2.

```
lr = 0.02
```


Then we can fit the head of our network.

```
learn.fit_one_cycle(5, slice(lr))
```

```
learn.save('stage-1-rn50')
```

...And fine-tune the whole model:

```
learn.unfreeze()
```

```
learn.lr_find()
learn.recorder.plot()
```

We get a different shape now. For the initial layers we find a point when loss starts to go up, and choose 10 times less (so 1e-05). For the final layers we choose 5 or 10 times less that the LR chosen for 1st stage (was 1e-02), as a rule of thumb.

```
learn.fit_one_cycle(5, slice(1e-5, lr/5))
```

```
learn.save('stage-2-rn50')
```

And now... we'll use **transfer learning**. We trained with 128x128 images, for speeding up the training but also to do transfer learning from our prior model knowledge to the original images (256x256). For the model it'll be a totally new dataset, and that will fix our possible overfitting, maybe improving our final accuracy.

Thanks to data block API we'll only change one part of the original chain, assigning the new dataset to the prior learner.

```
data = (src.transform(tfms, size=256)
        .databunch().normalize(imagenet_stats))

learn.data = data
data.train_ds[0][0].shape
```

We repeat the whole process over the same model using the new data!

```
learn.freeze()
```

```
learn.lr_find()
learn.recorder.plot()
```

We don't see a clear negative slope now. With this kind of shape we can take the point before it shoots up (5e-2), and choose something like 10 times less (5e-3). Or check 2e-4.

```
lr=5e-3
```

```
learn.fit_one_cycle(5, slice(lr))
```

```
learn.save('stage-1-256-rn50')
```

```
learn.unfreeze()
```

```
learn.fit_one_cycle(5, slice(1e-5, lr/5))
```

Terrific result!!!

```
learn.recorder.plot_losses()
```

```
learn.save('stage-2-256-rn50')
```

You won't really know how you're going until you submit to Kaggle, since the leaderboard isn't using the same subset as we have for training. But as a guide, 50th place (out of 938 teams) on the private leaderboard was a score of `0.930`.

```
learn.export()
```

```
test = ImageList.from_folder(path/'test-jpg').add(ImageList.from_folder(path/'test-jpg-additional'))
len(test)
```

```
learn = load_learner(path, test=test)
preds, _ = learn.get_preds(ds_type=DatasetType.Test)
```

```
thresh = 0.2
labelled_preds = [' '.join([learn.data.classes[i] for i,p in enumerate(pred) if p > thresh]) for pred in preds]
```

```
labelled_preds[:5]
```

```
fnames = [f.name[:-4] for f in learn.data.test_ds.items]
```

```
df = pd.DataFrame({'image_name':fnames, 'tags':labelled_preds}, columns=['image_name', 'tags'])
```
